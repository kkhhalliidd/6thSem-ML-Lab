```python
# Import required libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
# Load dataset
data = pd.read_csv("IRISS.csv")

# Fix column spacing issue (important)
data.columns = data.columns.str.strip()

# Features and target
X = data.drop("species", axis=1)
y = data["species"]

# Train-test split (80-20)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=68
)
```

```python
# 1. Default Random Forest (10 trees)

rf_default = RandomForestClassifier(n_estimators=10, random_state=42)
rf_default.fit(X_train, y_train)

y_pred_default = rf_default.predict(X_test)
default_score = accuracy_score(y_test, y_pred_default)

print("Accuracy with default (10 trees):", default_score)

# Confusion Matrix (default)
cm_default = confusion_matrix(y_test, y_pred_default)
print("\nConfusion Matrix (Default):\n", cm_default)
```

```python
# 2. Hyperparameter Tuning

best_score = 0
best_n = 0

for n in range(1, 101):
    rf = RandomForestClassifier(n_estimators=n, random_state=42)
    rf.fit(X_train, y_train)

    y_pred = rf.predict(X_test)
    score = accuracy_score(y_test, y_pred)

    if score > best_score:
        best_score = score
        best_n = n

print("\nBest Accuracy:", best_score)
print("Best Number of Trees:", best_n)
```

```python
# 3. Final Model (Best Trees)

rf_best = RandomForestClassifier(n_estimators=best_n, random_state=42)
rf_best.fit(X_train, y_train)

y_pred_best = rf_best.predict(X_test)

# Confusion Matrix (best)
cm_best = confusion_matrix(y_test, y_pred_best)
print("\nConfusion Matrix (Best Model):\n", cm_best)
```
