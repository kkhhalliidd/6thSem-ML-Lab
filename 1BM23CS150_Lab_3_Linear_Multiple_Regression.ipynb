```python
import pandas as pd
from sklearn.linear_model import LinearRegression

# Load dataset
df = pd.read_csv("hiring.csv")

# Clean column names
df.columns = df.columns.str.strip()

# Convert word numbers to numeric
word_to_num = {
    'zero':0,'one':1,'two':2,'three':3,'four':4,
    'five':5,'six':6,'seven':7,'eight':8,'nine':9,
    'ten':10,'eleven':11,'twelve':12
}

df['experience'] = df['experience'].replace(word_to_num)
df['experience'] = df['experience'].fillna(0)

df['test_score(out of 10)'] = df['test_score(out of 10)'].fillna(
    df['test_score(out of 10)'].mean()
)

df['interview_score(out of 10)'] = df['interview_score(out of 10)'].fillna(
    df['interview_score(out of 10)'].mean()
)

# Features and target
X = df[['experience','test_score(out of 10)','interview_score(out of 10)']]
y = df['salary($)']

# Train model
model = LinearRegression()
model.fit(X, y)

# Predict using DataFrame (removes sklearn warning)
candidate1 = pd.DataFrame([[2,9,6]], columns=X.columns)
candidate2 = pd.DataFrame([[12,10,10]], columns=X.columns)

print("Salary prediction (2yr exp,9 test,6 interview):",
      model.predict(candidate1)[0])

print("Salary prediction (12yr exp,10 test,10 interview):",
      model.predict(candidate2)[0])
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression

# Load dataset
df = pd.read_csv("1000_Companies.csv")

# Clean column names
df.columns = df.columns.str.strip()

# Convert State categorical variable → numerical
df = pd.get_dummies(df, columns=['State'], drop_first=True)

# Features and target
X = df.drop('Profit', axis=1)
y = df['Profit']

# Train model
model = LinearRegression()
model.fit(X, y)

# Prediction sample (Florida)
sample = pd.DataFrame(
    [[91694.48, 515841.3, 11931.24, 1, 0]],
    columns=X.columns
)

print("Predicted Profit:", model.predict(sample)[0])
```
